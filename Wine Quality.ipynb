## for best quality wine alcohol should be above 13 for medium quality wine alcohol should be above 10.50 and below this is worest wine quality

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import ExtraTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,precision_score
from sklearn.linear_model import LogisticRegression
```

```python
x=df[['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']]
y=df['quality']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=3)
```

```python
DTC=DecisionTreeClassifier()
KNC=KNeighborsClassifier()
SVClas=SVC()
ETC=ExtraTreeClassifier()
RFC=RandomForestClassifier()
Lr=LogisticRegression()
```

```python
DTC.fit(x_train,y_train)
```

```python
KNC.fit(x_train,y_train)
```

```python
SVClas.fit(x_train,y_train)
```

```python
ETC.fit(x_train,y_train)
```

```python
RFC.fit(x_train,y_train)
```

```python
Lr.fit(x_train,y_train)
```

## acurracy score on diffrent classification model on predicted test data

```python
y_predDTC=DTC.predict(x_test)
print("accuracy_score" ,accuracy_score(y_test,y_predDTC)*100)
print("precision_score" ,precision_score(y_test,y_predDTC)*100)
```

```python
y_predKNC=KNC.predict(x_test)
accuracy_score(y_test,y_predKNC)*100
print("accuracy_score" ,accuracy_score(y_test,y_predKNC)*100)
print("precision_score" ,precision_score(y_test,y_predKNC)*100)
```

```python
y_predSVC=SVClas.predict(x_test)
print("accuracy_score" ,accuracy_score(y_test,y_predSVC)*100)
print("precision_score" ,precision_score(y_test,y_predSVC)*100)
```

```python
y_predETC=ETC.predict(x_test)
print("accuracy_score" ,accuracy_score(y_test,y_predETC)*100)
print("precision_score" ,precision_score(y_test,y_predETC)*100)
```

```python
y_predRFC=RFC.predict(x_test)
print("accuracy_score" ,accuracy_score(y_test,y_predRFC)*100)
print("precision_score" ,precision_score(y_test,y_predRFC)*100)
```

```python
y_predLR=Lr.predict(x_test)
print("accuracy_score" ,accuracy_score(y_test,y_predLR)*100)
print("precision_score" ,precision_score(y_test,y_predLR)*100)
```

## We have select the Random Forest Classification with 92 percent of accuracy and and 82 percent of precision score and they wine quality is 0 and 1 which is imbalace data so precision score should be high not accuracy
