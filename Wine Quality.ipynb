```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import ExtraTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
x=df[['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']]
y=df['quality']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
DTC=DecisionTreeClassifier()
KNC=KNeighborsClassifier()
SVClas=SVC()
ETC=ExtraTreeClassifier()
RTC=RandomForestClassifier()
```

```python
DTC.fit(x_train,y_train)
```

```python
KNC.fit(x_train,y_train)
```

```python
SVClas.fit(x_train,y_train)
```

```python
ETC.fit(x_train,y_train)
```

```python
RTC.fit(x_train,y_train)
```

## acurracy score on diffrent classification model on predicted test data

```python
y_predDTC=DTC.predict(x_test)
accuracy_score(y_test,y_predDTC)*100
```

```python
y_predKNC=KNC.predict(x_test)
accuracy_score(y_test,y_predKNC)*100
```

```python
y_predSVC=SVClas.predict(x_test)
accuracy_score(y_test,y_predSVC)*100
```

```python
y_predETC=ETC.predict(x_test)
accuracy_score(y_test,y_predETC)*100
```

```python
y_predRTC=RTC.predict(x_test)
accuracy_score(y_test,y_predRTC)*100
```

```python
y_predRTC=RTC.predict(x_train)
accuracy_score(y_train,y_predRTC)*100
```
